```
from abc import ABC, abstractmethod

# 1. Abstracción
class Cancion(ABC):
    def __init__(self, titulo, artista, duracion):
        self.titulo = titulo
        self.artista = artista
        self.__duracion = duracion  # encapsulado

    @abstractmethod
    def reproducir(self):
        pass

    # 2. Encapsulamiento
    def obtener_duracion(self):
        return self.__duracion


# 3. Herencia
class CancionRock(Cancion):
    def reproducir(self):
        return f"Reproduciendo ROCK: {self.titulo} - {self.artista} "


class CancionAlternativa(Cancion):
    def reproducir(self):
        return f"Reproduciendo ALTERNATIVO: {self.titulo} - {self.artista} "


# Uso del programa
cancion1 = CancionRock("All the Small Things", "Blink-182", 2.5)
cancion2 = CancionAlternativa("Chau", "No Te Va Gustar", 4)

# Polimorfismo
print(cancion1.reproducir())
print(f"Duración: {cancion1.obtener_duracion()} min")

print(cancion2.reproducir())
print(f"Duración: {cancion2.obtener_duracion()} min")
```

```
from abc import ABC, abstractmethod

# 1. Abstracción
class Medicamento(ABC):
    def __init__(self, nombre, dosis_mg, via):
        self.nombre = nombre
        self.__dosis_mg = dosis_mg  # encapsulado
        self.via = via

    @abstractmethod
    def administrar(self):
        pass

    # 2. Encapsulamiento
    def obtener_dosis(self):
        return self.__dosis_mg


# 3. Herencia
class Antibiotico(Medicamento):
    def __init__(self, nombre, dosis_mg, via, espectro):
        super().__init__(nombre, dosis_mg, via)
        self.espectro = espectro

    # 4. Polimorfismo
    def administrar(self):
        return f"Administrando antibiótico {self.nombre} vía {self.via} para bacterias {self.espectro}"


class Analgesico(Medicamento):
    def __init__(self, nombre, dosis_mg, via, tipo_dolor):
        super().__init__(nombre, dosis_mg, via)
        self.tipo_dolor = tipo_dolor

    def administrar(self):
        return f"Administrando analgésico {self.nombre} vía {self.via} para dolor {self.tipo_dolor}"


# Uso del sistema
med1 = Antibiotico("Amoxicilina", 500, "oral", "Gram positivas")
med2 = Analgesico("Ibuprofeno", 400, "oral", "inflamatorio")

print(med1.administrar())
print(f"Dosis: {med1.obtener_dosis()} mg")

print(med2.administrar())
print(f"Dosis: {med2.obtener_dosis()} mg")
```

```
from abc import ABC, abstractmethod

# 1. Abstracción
class Animal(ABC):
    def __init__(self, nombre, edad, peso):
        self.nombre = nombre
        self.edad = edad
        self.__peso = peso  # encapsulado

    @abstractmethod
    def hacer_sonido(self):
        pass

    @abstractmethod
    def tratamiento(self):
        pass

    # 2. Encapsulamiento
    def obtener_peso(self):
        return self.__peso


# 3. Herencia
class Perro(Animal):
    def hacer_sonido(self):
        return f"{self.nombre} dice: ¡Guau! "

    # 4. Polimorfismo
    def tratamiento(self):
        return f"{self.nombre} necesita vacuna antirrábica"


class Gato(Animal):
    def hacer_sonido(self):
        return f"{self.nombre} dice: ¡Miau! "

    def tratamiento(self):
        return f"{self.nombre} necesita desparasitación"


# Uso del sistema
animal1 = Perro("Max", 3, 20)
animal2 = Gato("Luna", 2, 5)

print(animal1.hacer_sonido())
print(animal1.tratamiento())
print(f"Peso: {animal1.obtener_peso()} kg")

print(animal2.hacer_sonido())
print(animal2.tratamiento())
print(f"Peso: {animal2.obtener_peso()} kg")
```
